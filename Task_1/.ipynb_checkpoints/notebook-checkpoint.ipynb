```python
def replace_none_with_zero(lst):
    if lst is None:
        return [0, 0, 0]
    else:
        for i in range(len(lst)):
            if lst[i] is None:
                lst[i] = [0, 0, 0]
            elif isinstance(lst[i], list):
                lst[i] = replace_none_with_zero(lst[i])
        return lst
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation

# Assuming 'matrix' is defined somewhere in your code
points_array = np.array(matrix)[0]
print(points_array)
x = points_array[:, 0]
y = points_array[:, 1]
z = points_array[:, 2]

# Create the figure and axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Initialize empty scatter plot
sc = ax.scatter([], [], [])

# Set the initial view
ax.set_xlim(np.min(x), np.max(x))
ax.set_ylim(np.min(y), np.max(y))
ax.set_zlim(np.min(z), np.max(z))

# Function to update the plot for each frame
def update(frame):
    sc._offsets3d = (x[:frame], y[:frame], z[:frame])
    ax.set_title('Frame {}'.format(frame))

# Create the animation
ani = FuncAnimation(fig, update, frames=500, interval=10)

# If you want to save the animation to a file
ani.save('3d_animation.gif', writer='pillow', fps=60)

plt.show()
```
